```python
import pandas as pd

#membaca file CSV
file_path='/content/500_Person_Gender_Height_Weight_Index (1).csv'
data = pd.read_csv(file_path)

#menampilkan seluruh data
```

```python
#statistik berdasar pada gender
stats = data.groupby('Gender').agg({
    'Height': ['count', 'mean', 'max','min'],
    'Weight': ['count', 'mean', 'max','min'],
})
```

```python
#mengubah kolom menjadi data binary
data['Gender'] = data['Gender'].apply(lambda x: 1 if x == 'Male' else 0)

#menampilkan data setelah perubahan
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#plot BMI index berdasarkan dengan Gender
plt.figure(figsize=(8,6))
sns.countplot(x='Index', hue='Gender', data=data)
plt.title('BMI Index berdasarkan Gender')
plt.xlabel('BMI Index')
plt.ylabel('jumlah')
plt.legend(title='Gender')
plt.show()
```
