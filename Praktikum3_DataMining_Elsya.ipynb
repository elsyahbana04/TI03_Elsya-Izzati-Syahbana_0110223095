```python
# import dan tampilkan dataset
import pandas as pd
import numpy as np

df = pd.read_csv("/content/BL-Flickr-Images-Book.csv")
```

```python
#melihat datatype dan spesifik pada dataset
df.info()
```

```python
#mengecek apakah ada missing values
df.isnull().sum()
```

```python
#Dropping unnecessary columns atau
#hapus data yang tidak diperlukan karna keseluruhan data memiliki nilai kosng
columns_to_drop = ['Edition Statement', 'Corporate Author',
                   'Corporate Contributors', 'Former owner', 'Engraver']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
#menjadikan indentifier sebagai index agar memudahkan mengambil sebuah data
df.set_index('Identifier', inplace=True)
df
```

```python
# Renaming colums
df.rename(columns={
    'Place of Publication': 'Tempat_Penerbitan',
    'Date of Publication': 'Tahun_Penerbitan',
    'Title ': 'Judul',
    'Author': 'Penulis',
    'Contributors': 'Kontributor',
    'Issuance type': 'Jenis_Penerbitan',
    'Flickr URL': 'URL_Flickr',
    'Shelfmarks': 'Penanda_Rak'
}, inplace=True)

#Tampilkan DataFrame setelah perubahan nama kolom
df
```
